```python
import cv2
import numpy as np 
import pandas as pd
import scipy as sp
from scipy.signal.signaltools import correlate2d as c2d
from PIL import Image
import os
import warnings
warnings.simplefilter('ignore')
```

```python
path = 'L:/Videodata/dataset/received'
files = os.listdir(path)
files_dir = [f for f in files if os.path.isfile(os.path.join(path, f))]
print(len(files_dir))    # ['dir1', 'dir2']
```

```python
received = []
for i in files_dir:
    received.append(path+"/"+i)
print(len(received))
print(received[0])
```

```python
ds = [] 
for f in range (len(received)):
    if f >419:
        org = cv2.VideoCapture('L:/Videodata/dataset/original/evBD9ieSL7g.mp4')
    elif f >359:
        org = cv2.VideoCapture('L:/Videodata/dataset/original/dz7T4rJPy44.mp4')
    elif f >299:
        org = cv2.VideoCapture('L:/Videodata/dataset/original/cwayE8SzvvI.mp4')
    elif f >239:
        org = cv2.VideoCapture('L:/Videodata/dataset/original/9Qc_1JBJmj0.mp4')   
    elif f >179:
        org = cv2.VideoCapture('L:/Videodata/dataset/original/9JbdHUpGkdE.mp4')
    elif f >119:
        org = cv2.VideoCapture('L:/Videodata/dataset/original/8jTJXuxVmVc.mp4')
    elif f >59:
        org = cv2.VideoCapture('L:/Videodata/dataset/original/7AvbTPa2Tv0.mp4')    
    else:    
        org = cv2.VideoCapture('L:/Videodata/dataset/original/0GHpTnbnTZs.mp4')
    print(received[f])
    rec = cv2.VideoCapture(received[f])
    sum_distance_CORREL = 0
    sum_distance_CHISQR = 0
    sum_distance_INTERSECT = 0
    sum_distance_BHATTACHARYYA = 0
    org.set(cv2.CAP_PROP_POS_FRAMES, 0)
    rec.set(cv2.CAP_PROP_POS_FRAMES, 0)
    for i in range (int(org.get(cv2.CAP_PROP_FRAME_COUNT))):
        reto, frameo = org.read()
        retr, framer = rec.read()
        h1 = cv2.calcHist([frameo], [0], None, [256], [0, 256])
        h2 = cv2.calcHist([framer], [0], None, [256], [0, 256])
        sum_distance_CORREL += np.abs(cv2.compareHist(h1,h2,0))
        sum_distance_CHISQR += np.abs(cv2.compareHist(h1,h2,1))
        sum_distance_INTERSECT += np.abs(cv2.compareHist(h1,h2,2))
        sum_distance_BHATTACHARYYA += np.abs(cv2.compareHist(h1,h2,3))
    ds.append([sum_distance_CORREL,sum_distance_CHISQR,sum_distance_INTERSECT,sum_distance_BHATTACHARYYA])
    print(sum_distance_CORREL,sum_distance_CHISQR,sum_distance_INTERSECT,sum_distance_BHATTACHARYYA)
```

```python
name_df = pd.Series(received)
data_df = pd.Series(ds)
train = pd.concat([name_df,data_df],axis=1)
train.to_csv('frametrain.csv')
```
